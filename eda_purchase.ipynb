```python
# STEP 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Create directory for saving graphs
os.makedirs("graphs", exist_ok=True)

# STEP 2: Load Data
df = pd.read_csv(r"C:\Users\Admin\OneDrive\Documents\TopMentor\Decision Tree\iphone_purchase_records.csv")
print("Data Loaded. Shape:", df.shape)
df.head()
```

```python
# STEP 3: Basic Info
print("ðŸ”¹ Data Types:\n", df.dtypes)
print("ðŸ”¹ Null Values:\n", df.isnull().sum())
print("ðŸ”¹ Target Distribution:\n", df['Purchase Iphone'].value_counts())
```

```python
# STEP 4: Gender Distribution
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.savefig("graphs/gender_dist.png")
plt.show()
```

```python
# STEP 5: Age Distribution
sns.histplot(df['Age'], kde=True, bins=20)
plt.title("Age Distribution")
plt.savefig("graphs/age_dist.png")
plt.show()
```

```python
# STEP 6: Salary Distribution
sns.histplot(df['Salary'], kde=True, color='green', bins=20)
plt.title("Salary Distribution")
plt.savefig("graphs/salary_dist.png")
plt.show()
```

```python
# STEP 7: Purchase vs Gender
sns.countplot(data=df, x='Gender', hue='Purchase Iphone')
plt.title("Purchase by Gender")
plt.savefig("graphs/purchase_by_gender.png")
plt.show()
```

```python
# STEP 8: Purchase by Age & Salary (Boxplots)
sns.boxplot(data=df, x='Purchase Iphone', y='Age')
plt.title("Age vs Purchase")
plt.savefig("graphs/age_vs_purchase.png")
plt.show()

sns.boxplot(data=df, x='Purchase Iphone', y='Salary')
plt.title("Salary vs Purchase")
plt.savefig("graphs/salary_vs_purchase.png")
plt.show()
```
